[코드 gsm 9-1] 상관분석 데이터 준비

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')

# 리스트에 데이터 삽입하기
X = [1, 2, 3, 4, 5]
Y_linear = [2, 4, 6, 8, 10] # 선형 데이터

# 리스트를 데이터프레임으로 변환하기
data = {'X': X, 'Y_선형': Y_linear}
```

[코드 gsm 9-2] 그래프

```python
# 그래프 그리기
df = pd.DataFrame(data)
df.plot(x='X', y='Y_선형')
plt.show()
```

[코드 gsm 9-3] 피어슨 상관분석

```python
coef_p = df.corr(method='pearson')
print("피어슨 상관계수 (X와 Y_linear):")
print(coef_p)
```

[코드 gsm 9-4] 스피어만 상관분석

```python
# 스피어만 상관분석 수행하기 (X와 Y선형)
coef_s = df.corr(method='spearman')
print("\n스피어만 상관계수 (X와 Y_linear):")
print(coef_s)
```

[코드 gsm 9-5] 켄달 상관분석

```python
# 켄달 상관분석 수행하기 (X와 Y선형)
coef_k = df.corr(method='kendall')
print("\n켄달 상관계수 (X와 Y_linear):")
print(coef_k)
```

[코드 gsm 9-10] 상관관계 분석(A)

```python
import pandas as pd
import matplotlib.pyplot as plt

# 리스트에 데이터 삽입하기
X = [1, 2, 3, 4]
Y = [10, 9, 8, 7]

# 리스트를 데이터프레임으로 변환하기
data = {'XA': X, 'YA': Y}
df = pd.DataFrame(data)
df

# 그래프 그리기
df.plot(x='XA', y='YA')
plt.show()

# 피어슨 상관분석 수행하기 (A)
Acoef_p = df.corr(method='pearson')
print("피어슨 상관계수(A형):")
print(Acoef_p)

# 스피어만 상관분석 수행하기 (A)
Acoef_s = df.corr(method='spearman')
print("\n스피어만 상관계수 (A):")
print(Acoef_s)

# 켄달 상관분석 수행하기 (A)
Acoef_k = df.corr(method='kendall')
print("\n켄달 상관계수 (A):")
print(Acoef_k)
```

[코드 gsm 9-11] 상관관계 분석(B)

```python
import pandas as pd
import matplotlib.pyplot as plt

# 리스트에 데이터 삽입하기
X = [1, 2, 3, 4]
Y = [10, 20, 30, 30]

# 리스트를 데이터프레임으로 변환하기
data = {'XB': X, 'YB': Y}
df = pd.DataFrame(data)
df

# 그래프 그리기
df.plot(x='XB', y='YB')
plt.show()

# 피어슨 상관분석 수행하기 (B)
Acoef_p = df.corr(method='pearson')
print("피어슨 상관계수(B형):")
print(Acoef_p)

# 스피어만 상관분석 수행하기 (B)
Acoef_s = df.corr(method='spearman')
print("\n스피어만 상관계수 (B):")
print(Acoef_s)

# 켄달 상관분석 수행하기 (B)
Acoef_k = df.corr(method='kendall')
print("\n켄달 상관계수 (B):")
print(Acoef_k)
```
