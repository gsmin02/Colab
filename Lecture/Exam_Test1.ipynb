필요 라이브러리 import

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

데이터 준비

```python
키 = [160, 165, 170, 175, 180]
몸무게 = [55, 60, 65, 70, 75]
```

데이터프레임 생성

```python
df = pd.DataFrame({
  '키': 키,
  '몸무게': 몸무게
})
```

모델 학습용 데이터 준비

```python
X = df[['키']]    # 독립변수 (2차원 배열)
y = df['몸무게']  # 종속변수 (1차원 배열)
```

모델 학습

```python
model = LinearRegression()
model.fit(X, y)
```

새로운 키에 대한 몸무게 예측

```python
예측키 = [[190]]
예측몸무게 = model.predict(예측키)
```

예측 결과 출력

```python
print(f"키가 190cm일 때 예상 몸무게: {예측몸무게[0]:.1f}kg")
```

변환 후 출력

```python
예측키1 = [[190], [200]]
예측몸무게1 = model.predict(예측키1)

print(f"키가 190cm일 때 예상 몸무게: {예측몸무게1[0]:.1f}kg")
print(f"키가 200cm일 때 예상 몸무게: {예측몸무게1[1]:.1f}kg")
```

```python
기울기 = model.coef_[0]
절편 = model.intercept_
print(f"회귀식: 몸무게 = {기울기:.2f} × 키 + {절편:.2f}")
```

```python
결정계수 = model.score(X, y)
print(f"결정계수: {결정계수:.4f}")
```
