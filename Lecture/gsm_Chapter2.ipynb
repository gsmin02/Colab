```python
def sum_list(a):
  j = 0
  for i in a:
    j = j + i
  return j

list_a = [1, 2, 3, 4, 5]
print(sum_list(list_a)) # 호출
```

[코드 gsm 2-21]


```python
def gugudan(a):
  for i in a:
    for j in range(1, 10):
      print(str(i) + ' X ' + str(j) + ' = ' + str(i * j))
    print("---------------")

gugu = [3, 4, 5]
print(gugudan(gugu)) # 호출
```

```python
def check_even_num(x):
  if x % 2 == 0:
    return True
  else:
    return False

number = int (input('판별할 자연수를 입력하세요:'))
if(check_even_num(number)):
  print('짝수')
else:
  print('홀수')
```
