## 02 - 1 API 사용하기

웹 기반 API는 HTTP 프로토콜을 사용하지만 보통 HTML을 주고받지 않음  
일반적으로 CSV, JSON, XML 같은 파일을 사용함  
HTML의 경우 구조가 복잡하여 버그 가능성이 높고, 양이 많으면 복잡성이 더욱 증가하기 때문

```python
d = {"name":"데이터 분석"}
print(d['name'])
```

JSON 포맷은 파이썬 딕셔너리와 잘 호환되지만, 웹 기반 API로 데이터를 전달 시 딕셔너리가 아니라 텍스트로 전달해야 함  
파이썬의 json 패키지를 사용해 JSON 형식에 맞는 문자열로 변경함  
json.dumps() 함수의 ensure_ascii 매개변수를 False로 지정하여 16진수가 아닌 한글로 출력되게 설정

```python
import json

d_str = json.dumps(d, ensure_ascii=False)
print(d_str)
```

파이썬의 type() 함수를 이용해 정상적으로 문자열 객체로 변경되었는지 확인함

```python
print(type(d_str))
```

API로 전달 받는 경우에도 텍스트로 전달받기 때문에 문자열을 파이썬 딕셔너리로 변경해야 함  
json.loads() 함수를 이용해 텍스트를 딕셔너리로 변경

```python
d2 = json.loads(d_str)
print(d2['name'])
```

type() 함수를 이용해 정상적으로 변경되었는지 확인

```python
print(type(d2))
```

조금 더 복잡한 형태를 텍스트로 만들고 json.loads() 함수를 이용해 문자열을 딕셔너리로 변경함

```python
d3 = json.loads('{"name" : "데이터 분석", "author" : "김연아", "year" : 2024}')
print(d3['name'])
print(d3['author'])
print(d3['year'])
```

응용 중의 하나로 Value 안에 대괄호를 통해 리스트를 저장할 수 있음

```python
d3 = json.loads('{"name" : "데이터 분석", "author" : ["김연아", "박지성"], "year" : 2024}')
print(d3['author'][1])
```

세 겹 따옴표 ( " " " ) 를 통해 긴 문자열을 줄바꿈하여 입력할 수 있음

```python
d4_str = """
[
  {"name" : "데이터 분석", "author" : "김연아", "year" : 2024},
  {"name" : "머신러닝+딥러닝", "author" : "박지성", "year" : 2020}
]
"""
d4 = json.loads(d4_str)
print(d4[0]['name'])
```

분석이 가장 용이한 데이터 프레임으로 변경하기 위해 판다스의 read_json() 함수를 이용함  
  
※ 판다스에 String 형을 직접적으로 넣는 방식을 지원하지 않으므로 StringIO를 이용해 read_json 함수를 호출하는 방식으로 변경

```python
import pandas as pd
from io import StringIO

d4_io = StringIO(d4_str)
pd.read_json(d4_io)
```

일반 String 형이 아닌 파이썬 리스트도 DataFrame() 함수를 이용하여 변경 가능함

```python
pd.DataFrame(d4)
```

xml은 HTML의 복잡한 형태를 컴퓨터와 사람이 모두 읽고 쓰기 편한 문서 포맷을 위해 고안됨  
엘리먼트(element)들이 계층 구조를 이루며, 태그를 통해 표현할 수 있음  
이 때 부모 엘리먼트 혹은 부모 노드와 자식 엘리먼트라고 부름

```python
x_str = """
<book>
  <name>혼자 공부하는 데이터 분석</name>
  <author>홍길동</author>
  <year>2024</year>
</book>
"""
```

xml.etree.ElementTree 모듈의 fromstring() 함수를 통해 문자열을 XML로 변환 가능

```python
import xml.etree.ElementTree as et
book = et.fromstring(x_str)
```

type() 함수를 통해 정상적으로 변환되었는지 확인

```python
print(type(book))
```

xml 객체의 tag 속성을 출력하면 엘리먼트 이름을 쉽게 확인할 수 있음

```python
print(book.tag)
```

book 객체를 리스트로 변환하여 자식 엘리먼트를 구할 수 있음  
간단히 list() 함수를 통해 분류할 수 있음

```python
book_childs = list(book)
print(book_childs)
```

각 엘리먼트들의 값을 알기위해 text 속성으로 내용을 확인할 수 있음

```python
name, author, year = book_childs
print(name.text)
print(author.text)
print(year.text)
```

위 방법으로 텍스트를 출력할 수 있지만, XML의 경우 **자식 엘리먼트 순서가 항상 일정하다는 것을 보장하지 않음**  
따라서 findetext() 메서드를 사용하여 자동으로 텍스트로 변환할 수 있음

```python
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')
print(name)
print(author)
print(year)
```

조금 더 복잡한 형태의 XML은 다음과 같은 방식으로 엘리먼트를 감싸는 부모 엘리먼트를 만들어서 표현 가능함

```python
x2_str = """
<books>
  <book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>홍길동</author>
    <year>2024</year>
  </book>
  <book>
    <name>혼자 공부하는 딥러닝</name>
    <author>김연아</author>
    <year>2022</year>
  </book>
</books>
"""
```

x2_str의 부모 엘리먼트는 <book>이 아니라 <books>가 된 것을 확인하기 위해 fromstring() 함수를 사용

```python
books = et.fromstring(x2_str)
print(books.tag)
```

여러 개의 자식 엘리먼트를 확인하기 위한 방법 중 findall() 메서드와 for 문을 조합하는 방식이 있음  
동일한 이름을 가진 여러 개의 자식 엘리먼트를 찾는 경우 편리하게 사용할 수 있는 방식

```python
for book in books.findall('book'):
  name = book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')
  print(name)
  print(author)
  print(year)
  print() # 구분 기호
```

JSON처럼 XML을 바로 변형하기 위해서는 판다스의 read_xml() 함수를 통해 바로 변환이 가능함  
단, 판다스 1.3.0 이상부터 지원함  
※ 판다스에 String 형을 직접적으로 넣는 방식을 지원하지 않으므로 StringIO를 이용해 read_xml 함수를 호출하는 방식으로 변경

```python
x2_io = StringIO(x2_str)
pd.read_xml(x2_io)
```

검색 결과 페이지를 requests.get() 함수를 통해 HTML을 가져올 수 있음

```python
import requests
isbn = 9791190090018 # 책의 ISBN
url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
r = requests.get(url.format(isbn))
```

응답 객체를 사용해 검색 결과를 HTML로 출력

HTML 페이지를 통해 상세 페이지 링크를 찾고 **뷰티플수프**를 사용해 데이터를 가져옴

```python
from bs4 import BeautifulSoup
```

객체의 첫 번째 매개변수는 파싱할 HMTL 문서이고 두 번째는 파싱에 사용할 파서다.  
파서는 입력 데이터를 받아 데이터 구조를 만드는 라이브러리를 의미하며, 이 과정을 파싱이라고 함

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

이제 soup 객체에게 찾고 있는 a 태그의 위치를 찾아야 함  
태그 위치는 soup 객체의 find() 메서드를 사용하며, attrs 매개변수에 속성을 딕셔너리로 지정하면 됨

```python
prd_link = soup.find('a', attrs={'class':'gd_name'})
```

추출한 객체를 print() 함수로 출력 시 태그 안에 포함된 HTML을 출력함

```python
print(prd_link)
```

soup.find()로 찾은 HTML 객체는 딕셔너리처럼 사용해 태그 안의 속성을 참조할 수 있음  
이 방법으로 링크 주소인 href 속성의 값을 얻을 수 있음

```python
print(prd_link['href'])
```

위 같은 방식으로 뷰티플수프를 사용해 Yes24의 검색 결과 페이지에서 상세 페이지로 넘어가는 링크를 연결할 수 있음

이제 상세 페이지 주소를 만들어 requests.get() 함수를 호출함

```python
url = 'http://www.yes24.com'+prd_link['href']
r = requests.get(url)
```

r을 통해 가져온 HTML을 출력하면

```python
print(r.text)
```

이제 쪽수를 찾기 위해 다시 HTML 개발자 도구로 분석하여 쪽수가 위치한 곳을 찾음  
찾은 경우 뷰티플수프 객체를 생성하고 find() 메서드로 품목정보 태그를 찾아서 출력

```python
soup = BeautifulSoup(r.text, 'html.parser')
prd_detail = soup.find('div', attrs={'id':'infoset_specific'})
print(prd_detail)
```

품목정보 테이블의 행을 하나씩 검사해서 '쪽수, 무게, 크기'에 해당하는 <tr> 태그를 찾아야 하는 경우  
find_all() 메서드를 사용하여 특정 HTML 태그를 모두 찾아서 리스트로 반환한다

```python
prd_tr_list = prd_detail.find_all('tr')
print(prd_tr_list)
```

태그를 리스트로 추출하고 for 문으로 리스트를 순회하면서 특정 텍스트를 찾아서 저장함  
가져오기 위해 get_text() 메서드를 사용

```python
for tr in prd_tr_list:
  if tr.find('th').get_text() == '쪽수, 무게, 크기':
    page_td = tr.find('td').get_text()
    break
```

잘 되었는지 확인

```python
print(page_td)
```

이제 split() 메서드를 이용해서 필요한 쪽수만 걸러냄

```python
print(page_td.split()[0])
```

이제 전체 도서의 쪽수를 구하여 저장할 수 있음  
데이터프레임은 한 행씩 순차적으로 처리하는데 최적화되어 있지 않아서 for문으로 반복하는 것은 매우 비효율적임  
  
다음과 같은 순서로 전체 도서 쪽수를 구할 수 있음

3. 뷰티플수프로 HTML을 파싱
4. 뷰티플수프의 find() 메서드로 \<a> 태그를 찾아 상세 페이지 URL을 추출
5. requests.get() 함수로 다시 상세 페이지의 HTML을 가져옴
6. 뷰티플수프로 HTML을 파싱함
7. 뷰티플수프의 find() 메서드로 '품목정보' \<div> 태그 찾기
8. 뷰티플수프의 find_all() 메서드로 '쪽수'가 들어이쓴 \<tr> 태그 찾기
9. 찾은 테이블의 행에서 get_text() 메서드로 \<td> 태그에 들어 있는 '쪽수'를 가져옴



```python
def get_page_cnt(isbn):
  url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
  r = requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser')
  prd_info = soup.find('a', attrs={'class':'gd_name'})
  if prd_info == None:
    return ''
  url = 'http://www.yes24.com'+prd_info['href']
  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')
  prd_detail = soup.find('div', attrs={'id':'infoset_specific'})
  prd_tr_list = prd_detail.find_all('tr')
  for tr in prd_tr_list:
    if tr.find('th').get_text() == '쪽수, 무게, 크기':
      return tr.find('td').get_text().split()[0]
  return ''
```

새로 만든 get_page_cnt2() 함수를 apply() 메서드에 사용함  
각 행에 함수를 적용해야 하므로 axis 매개변수를 1로 지정함  
0인 경우 각 열에 대해 함수를 적용함
