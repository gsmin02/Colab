[코드 gsm 7-11] 네트워크 그래프 생성

```python
!pip install networkx
import networkx as nx

G = nx.DiGraph()
G.add_nodes_from(['A','B','C','D','E'])
G.add_edges_from([('A','B'), ('B','C'), ('C','D')])

nx.draw(G, with_labels=True)
```

[코드 gsm 7-12] 노드와 선 추가 및 제거

```python
#노드 추가 및 제거하기
g1 = nx.Graph()
g1.add_node('David')
g1.add_node('John')
g1.add_node('Mary')

#선 추가 및 제거하기
g1.add_edge('David', 'John')
g1.add_edge('David', 'Mary')
g1.remove_edge('Mary', 'David')

nx.draw(g1, with_labels=True, font_weight='bold')
```

[코드 gsm 7-13] 가중치 표시

```python
g3 = nx.DiGraph()
g3.add_weighted_edges_from([(1, 2, 10), (2, 3, 20)])
g3.add_edge(1, 3, weight=30)
pos = nx.spring_layout(g3)

nx.draw(g3, pos=pos, with_labels=True)
labels = nx.get_edge_attributes(g3, 'weight')
nx.draw_networkx_edge_labels(g3, pos, edge_labels=labels);
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#데이터를 데이터프레임 형태로 읽어오기
df = pd.read_csv('temp_gsm.csv', encoding='cp949')
df.head(3)
```

[코드 gsm 7-20] 특수 문자가 포함된 열 이름을 변경

```python
df2.rename(columns={'최저기온(°C)':'min_temp'}, inplace=True)
df2.rename(columns={'평균기온(°C)':'avg_temp'}, inplace=True)
df2.rename(columns={'최고기온(°C)':'max_temp'}, inplace=True)
df2.info()
```

[코드 gsm 7-21] 꺾은선 그래프 생성

```python
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('서울시 2024년도 기온 변화')

plt.plot(range(1,len(df2)+1), df2['max_temp'], label='최고기온', c='r')
plt.plot(range(1,len(df2)+1), df2['avg_temp'], label='평균기온', c='y')
plt.plot(range(1,len(df2)+1), df2['min_temp'], label='최저기온', c='b')
plt.xlabel('경과일')
plt.ylabel('기온')
plt.legend()
plt.show()
```

[코드 gsm 7-22] 일시 데이터를 날짜형으로 변경

[코드 gsm 7-30] 워드클라우드 생성

```python
#텍스트 분석하기
with open('대한민국헌법.txt', 'r', encoding='utf-8') as f: text = f.read()
okt = Okt()
nouns = okt.nouns(text)
words = [n for n in nouns if len(n)>1]
c = Counter(words)

#워드클라우드 생성하기
wc = WordCloud(font_path='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', width=400, height=400, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
wc.to_file('헌법_워드클라우드_gsm.png')
```
