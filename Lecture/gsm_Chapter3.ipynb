[코드 3-6] CSV 파일 작성

```python
#csv 라이브러리 가져오기
import csv
#CSV 파일 불러오기
f = open('writegsm.csv', 'w', encoding='cp949', newline='')
#CSV 파일 작성할 준비하기
wr = csv.writer(f)
#데이터를 한 행씩 작성하기
wr.writerow(['ID', '이름', '상징색', '취미', '특징'])
wr.writerow(['001', '뽀로로', '파랑색', '낚시', '펭귄'])
wr.writerow(['002', '에디', '주황색', '과학실험', '사막여우'])
wr.writerow(['003', '크롱', '초록색', '눈싸움', '공룡'])
wr.writerow(['004', '루피', '분홍색', '요리', '비버'])
wr.writerow(['005', '패티', '보라색', '운동', '펭귄'])

#f 변수의 자원 반환하기
f.close()
```

[코드 gsm 3-7] CSV 파일에 데이터 추가

[코드 gsm 3-9] 엑셀 파일 업로드

```python
#openpyxl 라이브러리 설치하기
!pip install openpyxl
#openpyxl 라이브러리 가져오기
import openpyxl

wb = openpyxl.load_workbook('charactersgsm.xlsx')
```

[코드 gsm 3-10] 특정 열 출력

```python
print(wb.sheetnames)
```

[코드 gsm 3-11] 특정 위치의 데이터 읽기

[코드 gsm 3-12] 새 워크북 생성

```python
wb = openpyxl.Workbook()
```

[코드 gsm 3-13] 새 워크시트 생성

```python
#새 워크시트 생성하기
wb.create_sheet('Sheet2')
#시트 이름 출력하기
print(wb.sheetnames)
```

[코드 gsm 3-14] 워크시트 이름 변경

```python
sheet1 = wb['Sheet']
sheet2 = wb['Sheet2']
sheet1.title = '캐릭터 명단'
sheet2.title = '인기도 조사'
print(wb.sheetnames)
```

[코드 gsm 3-15] 워크시트 이름 변경

```python
sheet2['B1'] = '인기도 조사결과'
print(sheet2['B1'].value)
```

[코드 gsm 3-16] 워크시트 복제

```python
copysheet = wb.copy_worksheet(sheet2)
print(wb.sheetnames)
#복세 시트의 이름 변경하기
copysheet.title = 'copy'
print(wb.sheetnames)
print(copysheet['B1'].value)
```

[코드 gsm 3-17] 워크시트 삭제

```python
del wb['인기도 조사']
print(wb.sheetnames)
```
